## Retrieving documents :

```python
def gen_pairs(dataset):
    ds = dataset.irds_ref()
    docsstore = ds.docs_store()
    queries = {q.query_id: q for q in ds.queries_iter()}
    while True:
      for doc_pair in ds.docpairs_iter():
        yield onir_pt.TrainPair(doc_pair.query_id,queries[doc_pair.query_id].text,
                                doc_pair.doc_id_a,docsstore.get(doc_pair.doc_id_a).text,
                                doc_pair.doc_id_b,docsstore.get(doc_pair.doc_id_b).text)
```

```python
model = None
trec_data = True
limit = 100
def BM25_BERT(dataset,index,trec_data):
      train_dataset = pt.get_dataset("irds:msmarco-passage/train") # ici je dois utiliser le même data set qu'eux

      model_path = './bert_trec'
      if not os.path.isfile(model_path):
        max_iteration = 5
        learning_rate = 1e-5
        batch_size = 2
        pre_validate = False
        patience = 50
        bert = onir_pt.reranker('vanilla_transformer',
                                'bert',
                                vocab_config={'train': True},
                                config={'max_train_it':max_iteration,
                                        'learning_rate': learning_rate,
                                        'batch_size': batch_size,
                                        'pre_validate': pre_validate,
                                        'patience':patience})
        if trec_data:
            bm25 = pt.BatchRetrieve(index,
                                    wmodel="BM25",
                                    verbose=True) % limit >> pt.text.get_text(dataset, 'text')

            validation_run = bm25(dataset.get_topics())

            bert.fit(va_run=validation_run,
                      va_qrels=dataset.get_qrels(),
                      tr_pairs=gen_pairs(train_dataset))
        else:
            retrieval_model = pt.BatchRetrieve(index, wmodel="BM25") % limit >> pt.text.get_text(train_dataset, 'text') >> bert
            retrieval_model.fit(
                train_dataset.get_topics(),
                train_dataset.get_qrels(),
                dataset.get_topics(),
                dataset.get_qrels())
        bert.to_checkpoint(model_path)

      bert = onir_pt.reranker.from_checkpoint(model_path)
      model = pt.BatchRetrieve(index, wmodel="BM25") % limit >> pt.text.get_text(dataset, 'text') >> bert
```
