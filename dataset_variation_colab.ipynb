```python
import string
import re

def suppression_ponctuation(text):
    punc = string.punctuation
    punc += '\n\r\t'
    text = text.translate(str.maketrans(punc, ' ' * len(punc)))
    text = re.sub('( )+', ' ', text)
    text = text.strip()

    return text.lower()
```

```python
datasets = ['antique/train/split200-valid' ]
original_queries = []
ids = []
category = []
variation = []
transformed_queries = []
dasaset_col = []

for dataset_name in datasets :

  dataset = ir_datasets.load(dataset_name)
  queries_id = [q[0] for q in dataset.queries_iter()]
  queries = [q[1].lower() for q in dataset.queries_iter()]

  #Suppression de la ponctuation
  delet_ponc_queries = []
  for q in queries :
    delet_ponc_queries.append(suppression_ponctuation(q))

  # Misspelling Category
  transformed_queries_1 = Misspelling.NeighbCharSwap(delet_ponc_queries)
  transformed_queries_2 = Misspelling.RandomCharSub (delet_ponc_queries)
  transformed_queries_3 = Misspelling.QWERTYCharSub(delet_ponc_queries)

  # Naturality Category
  transformed_queries_4 = Naturality.RemoveStopWords(delet_ponc_queries)
  transformed_queries_5 = Naturality.T5DescToTitle(delet_ponc_queries)

  # Ordering Category
  transformed_queries_6 = Ordering.RandomOrderSwap(delet_ponc_queries)

  # Paraphrasing Category
  transformed_queries_7 = Paraphrasing.BackTranslation(delet_ponc_queries)
  transformed_queries_8 = Paraphrasing.T5QQP (delet_ponc_queries)
  transformed_queries_9 = Paraphrasing.WordEmbedSynSwap(delet_ponc_queries)
  transformed_queries_10 = Paraphrasing.WordNetSynSwap(delet_ponc_queries)


  trans_q =  transformed_queries_1 +\
    transformed_queries_2  +\
    transformed_queries_3 + \
    transformed_queries_4 +  \
    transformed_queries_5 +  \
    transformed_queries_6 + \
    transformed_queries_7 + \
    transformed_queries_8  +\
    transformed_queries_9 + \
    transformed_queries_10

  transformed_queries.extend(trans_q)
  original_queries = delet_ponc_queries * 10
  ids= queries_id * 10
  # for i , q in enumerate(queries) :
  #   original_queries.extend([q] * 10)
  #   ids.extend([queries_id[i]] * 10)
  category = ["Misspelling"] * 3* len(queries)+ ["Naturality"] * 2* len(queries) + ["Ordering"] * len(queries) + ["Paraphrasing"] * 4* len(queries)
  variation= ["NeighbCharSwap"]* len(queries) + ["RandomCharSub"]* len(queries) + ["QWERTYCharSub"]* len(queries) + ["RemoveStopWords"]* len(queries) + ["T5DescToTitle"]* len(queries) + ["RandomOrderSwap"]* len(queries) + ["BackTranslation"]* len(queries) + ["T5QQP"]* len(queries) + ["WordEmbedSynSwap"] * len(queries) + ["WordNetSynSwap"]* len(queries)
  dasaset_col.extend([dataset_name] * 10 * len(queries))

print(len(dasaset_col))
print(len(ids))
print(len(original_queries))
print(len(transformed_queries))
print(len(category))
print(len(variation))

data = {
    "dataset" : dasaset_col,
    "query_id" : ids,
    "query_original" :  original_queries,
    "query_variation" : transformed_queries ,
    "category" : category,
    "variation" : variation
}
df = pd.DataFrame(data)
df['query_variation'] = df['query_variation'].apply(suppression_ponctuation)

from google.colab import files
csv_path = "Antique_variation.csv"
df.to_csv(csv_path, index=False)

files.download(csv_path)
```
