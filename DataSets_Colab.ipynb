# Création du dataframe pour stocker les transformations des requetes

```python
import Misspelling
import Naturality
import Ordering
import Paraphrasing

import pandas as pd
```

```python
datasets = ['msmarco-passage/trec-dl-2019/judged' , 'antique/train/split200-valid']


def generate_query_transformation(datasets , folder_path) :
  original_queries = []
  ids = []
  category = []
  variation = []
  transformed_queries = []
  dasaset_col = []

  for dataset_name in datasets :

    dataset = ir_datasets.load(dataset_name)
    queries_id = [q[0].lower() for q in dataset.queries_iter()]
    queries = [q[1].lower() for q in dataset.queries_iter()]

    # Misspelling Category
    transformed_queries_1 = Misspelling.NeighbCharSwap(queries)
    transformed_queries_2 = Misspelling.RandomCharSub (queries)
    transformed_queries_3 = Misspelling.QWERTYCharSub(queries)

    # Naturality Category
    transformed_queries_4 = Naturality.RemoveStopWords(queries)
    transformed_queries_5 = Naturality.T5DescToTitle(queries)

    # Ordering Category
    transformed_queries_6 = Ordering.RandomOrderSwap(queries)

    # Paraphrasing Category
    transformed_queries_7 = Paraphrasing.BackTranslation(queries)
    transformed_queries_8 = Paraphrasing.T5QQP (queries)
    transformed_queries_9 = Paraphrasing.WordEmbedSynSwap(queries)
    transformed_queries_10 = Paraphrasing.WordNetSynSwap(queries)

    trans_q =  transformed_queries_1 +\
      transformed_queries_2  +\
      transformed_queries_3 + \
      transformed_queries_3 + \
      transformed_queries_4 +  \
      transformed_queries_5 +  \
      transformed_queries_6 + \
      transformed_queries_7 + \
      transformed_queries_8  +\
      transformed_queries_9 + \
      transformed_queries_10

    transformed_queries.extend(trans_q)

    for i , q in enumerate(queries) :
      original_queries.extend([q] * 10)
      ids.extend([queries_id[i]] * 10)
      category.extend(["Misspelling"] * 3 + ["Naturality"] * 2 + ["Ordering"] + ["Paraphrasing"] * 4)
      variation.extend(["NeighbCharSwap" , "RandomCharSub" , "QWERTYCharSub" , "RemoveStopWords" , "T5DescToTitle" , "RandomOrderSwap" , "BackTranslation" , "T5QQP" , "WordEmbedSynSwap" , "WordNetSynSwap"])
    dasaset_col.extend([dataset_name] * 10 * len(queries))
  data = {
      "dataset" : dasaset_col,
      "query_id" : ids,
      "query_original" :  original_queries,
      "query_variation" : transformed_queries ,
      "category" : category,
      "variation" : variation
  }
  df = pd.DataFrame(data)
  df.to_csv(folder_path+"/queries_variation.csv")
  return df
```
